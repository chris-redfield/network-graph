```python
df_socio = pd.read_csv("Socio.csv", sep=";")
```

```python
print(df_socio.shape)
df_socio.head()
```

```python
df_socio.GRUPO.value_counts()
```

```python
grupo_adapter = {"Sócio":2,"Sócio e Dirigente":3,"Fornecedor":4,"Conjuge":5,"Dirigente":6,"Fornecedor do Conjuge":7}
```

### Prepara para o join

```python
df_socio['SOCIO_BACKUP'] = df_socio.SOCIO
df_socio.SOCIO = df_socio.ID_SOCIO
```

### remove duplicados

```python
df_socio = df_socio.drop_duplicates(subset='NO_SOCIO', keep="last")
```

```python
df_joined = pd.merge(df_bb, df_socio, on=['SOCIO'],how='inner')
```

```python
#df_join = df_bb.set_index('SOCIO').join(df_socio.set_index('ID_SOCIO'))
```

```python
df_joined
```

### Cria json

```python
data = {}
data['nodes'] = [{
      "id": int(df_joined.ID_FORNECEDOR[0]),
      "name": df_joined.NO_FORNECEDOR[0],
      "group": 1
    }]

data['links'] = []

for index, row in df_joined.iterrows():
    data['nodes'].append({"id":int(row.SOCIO),"name":row.NO_SOCIO,"group":grupo_adapter[row.GRUPO]})
    data['links'].append({"source":int(row.SOCIO),"target":int(df_joined.ID_FORNECEDOR[0])})
```

### preview para checar o arquivo

```python
len(data['nodes'])
```

```python
json_data = json.dumps(data)
json_data
```

### Busca pelos socios dos socios

```python
socios_empresa = df_joined.SOCIO.values
socios_empresa
```

```python
#recarrega df_socio pois ele foi alterado
df_socio = pd.read_csv("Socio.csv", sep=";")
```

```python
df_sociodosocio = df_socio.loc[df_socio['SOCIO'].isin(socios_empresa)]
print("shape após busca de socios:", df_sociodosocio.shape)

# pega só os que não estão se auto referenciando
df_sociodosocio = df_sociodosocio.loc[df_sociodosocio.ID_SOCIO != df_sociodosocio.SOCIO]
print("shape após remoção das auto referências:", df_sociodosocio.shape)

# remove id_SOCIO = 1.0 (banco do brasil)
df_sociodosocio = df_sociodosocio.loc[df_sociodosocio.ID_SOCIO != 1.0]
print("shape após de auto referências à empresa inicial:", df_sociodosocio.shape)
```

```python
for index, row in df_sociodosocio.iterrows():
    if(int(row.SOCIO) in socios_empresa):
        if(row.ID_SOCIO not in socios_empresa):
            print("elemento ", row.NO_SOCIO, "não está na lista")
            data['nodes'].append({"id":int(row.ID_SOCIO),"name":row.NO_SOCIO,"group":grupo_adapter[row.GRUPO]})
            socios_empresa = np.append(socios_empresa, row.ID_SOCIO)
        data['links'].append({"source":int(row.ID_SOCIO),"target":int(row.SOCIO)})
```

```python
json_data = json.dumps(data)
json_data
```

```python
with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```
