```python
df_bb = df_empresa_socio.loc[df_empresa_socio.ID_FORNECEDOR==191]
df_bb.shape
```

```python
df_socio = pd.read_csv("Socio.csv", sep=";")
```

```python
print(df_socio.shape)
df_socio.head()
```

```python
df_socio.loc[df_socio.ID_SOCIO==11065391846]
```

### Prepara para o join

```python
df_socio.SOCIO = df_socio.ID_SOCIO
```

### remove duplicados

```python
df_socio = df_socio.drop_duplicates(subset='NO_SOCIO', keep="last")
```

```python
df_joined = pd.merge(df_bb, df_socio, on=['SOCIO'],how='inner')
```

```python
#df_join = df_bb.set_index('SOCIO').join(df_socio.set_index('ID_SOCIO'))
```

```python
df_joined
```

### Cria json

```python
data = {}
data['nodes'] = [{
      "id": int(df_joined.ID_FORNECEDOR[0]),
      "name": df_joined.NO_FORNECEDOR[0],
      "group": 1
    }]

data['links'] = []

for index, row in df_joined.iterrows():
    data['nodes'].append({"id":int(row.SOCIO),"name":row.NO_SOCIO,"group":2})
    data['links'].append({"source":int(row.SOCIO),"target":int(df_joined.ID_FORNECEDOR[0])})
```

### preview para checar o arquivo

```python
json_data = json.dumps(data)
json_data
```

```python
with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```
